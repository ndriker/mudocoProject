```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("nesterov_250_results.txt", "r") as f:
    arr = [[float(x) for x in line.split(":")] for line in f]
res = np.array(arr)
box = np.ones(10) / 10
#y = np.convolve(res[1:,1],box,mode='same')
y = res[1:,1]
```

```python
plt.plot(res[1:,0], y)
degree = 3
z = np.polyfit(res[1:,0], y, degree)
p = np.poly1d(z)
xp = np.linspace(0, np.max(res[1:,0]))
plt.plot(xp, p(xp), '--r', linewidth=3)
```
